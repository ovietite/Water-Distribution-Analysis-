**Prepare the Data**

```python
# Fill missing values in 'ACTUAL_MIN_NIGHT_FLOW' with the median value
median = df['ACTUAL_MIN_NIGHT_FLOW'].median()
df['ACTUAL_MIN_NIGHT_FLOW'].fillna(median, inplace=True)
```

**Model Selection**

Machine learning can be valuable for identifying anomalous water flow patterns that may suggest potential leaks. However, the current dataset lacks a clear target variable (such as a label indicating "leak" or "no leak") required for a supervised learning approach like classification. In this context, unsupervised methods, such as clustering (e.g., K-Means or Isolation Forest), could be effective for grouping DMAs with similar flow characteristics. DMAs that deviate significantly from these clusters can then be flagged for further investigation to detect possible leaks.  Isolation Forest isolates outliers effectively, making it ideal for identifying rare events such as leaks.




```python
from sklearn.ensemble import IsolationForest

# features to be used in the model
features = df[['ACTUAL_MIN_NIGHT_FLOW', 'MIN_NIGHT_FLOW', 'CENTROID_X', 'CENTROID_Y']]


iso_forest = IsolationForest(contamination=0.01, random_state=42)  #assumption that 1% of the data cotains anomalies
iso_forest.fit(features)

#(-1 for anomaly, 1 indicates normal)
df['anomaly'] = iso_forest.predict(features)
#store anomalies in a df
anomalies = df[df['anomaly'] == -1]

# Step 6: Plot anomalies on a scatter plot (Actual vs Min Night Flow)
plt.figure(figsize=(10, 6))
plt.scatter(df['ACTUAL_MIN_NIGHT_FLOW'], df['MIN_NIGHT_FLOW'], label='Normal', alpha=0.6, color='blue')
plt.scatter(anomalies['ACTUAL_MIN_NIGHT_FLOW'], anomalies['MIN_NIGHT_FLOW'], label='Anomalies', color='red', alpha=0.8)
plt.title('Anomaly Detection - Isolation Forest')
plt.xlabel('ACTUAL_MIN_NIGHT_FLOW')
plt.ylabel('MIN_NIGHT_FLOW')
plt.legend()
plt.show()

# Step 7: Output anomalies for further inspection
anomalies.head()
```

**Model Evaluation:**

Since we lack labeled data to indicate leaks, it is difficult to apply traditional supervised metrics such as precision or recall. Instead, the evaluation of the results is based on visual confirmation and domain-specific anomaly thresholds. Using Isolation Forest with an assumed contamination rate of 1%, the model successfully identified outliers that significantly deviate from normal flow patterns. Notable anomalies were detected in DMAs like AW_7812, AW_2708, and AW_9935, which exhibited higher-than-normal night flows, potentially indicating leaks.


**Limitations:**

*   *No Labelled Data:* Without labeled data on actual leaks, it's impossible to directly validate the model’s performance. The anomalies detected are speculative and need manual validation or further domain expertise. we assumed that 1% of the data contains anomalies.
*   *Data Quality:* There are some anomalies in the data, like negative flow values, which could affect the model’s reliability as well as some missing data. This could be dealt with adjustment/tweaks to the pipelines that collect the data.
*   *Simplistic Feature Set:* The model only considers night flow and geographic data. More features could improve accuracy of the model.



**Recommendations for Future Work:**

*   *Incorporate Labeled Data:* Collect labeled data from real-world leak reports and use it for supervised learning models like classification, which could provide more accurate predictions of leaks.
*   *Expand Features:* Add more contextual features to the model to improve the detection of true leaks.
*   *Anomaly Validation:* Investigate the flagged DMAs in the real world to assess the accuracy of the anomaly detection and refine the model accordingly.
